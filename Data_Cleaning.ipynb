# Remove Outlier using IQR

```python
sns.boxplot(x="CoapplicantIncome",data=df)
plt.show()
```

```python
q1=df["CoapplicantIncome"].quantile(0.25) # first quartile
q1
```

```python
q3=df["CoapplicantIncome"].quantile(0.75) # third quartile
q3
```

```python
IQR=q3-q1   # inter quartile range(IQR)
IQR
```

```python
min_range=q1-(1.5*IQR)  # minimum range
max_range=q3+(1.5*IQR)  # maximum range
min_range,max_range
```

```python
new_df=df[df["CoapplicantIncome"]<=max_range]  #maxmimum range se km jo data ha wo mily ga.bki remove ho jy ga
```

```python
new_df.shape   # here outliers remove from dataset
```

```python
sns.boxplot(x="CoapplicantIncome",data=new_df)  # here we can see that outlier remove from dataset.(few outlier is present but it is not effect)
plt.show()
```

# ***Remove Outlier Using Z Score***

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df=pd.read_csv("loan.csv")
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
sns.boxplot(x="CoapplicantIncome",data=df)
plt.show()
```

```python
sns.distplot(df["CoapplicantIncome"])
plt.show()
```

```python
z_score=(df["CoapplicantIncome"]-df["CoapplicantIncome"].mean())/(df["CoapplicantIncome"].std())   #  z_score =x-mean(x)/standard deviation
```

```python
z_score
```

```python
df["z_score"]=z_score    # add now column z_score in dataframe & then assing value of z_score to the column
```

```python
df
```

```python
df[df["z_score"]<3]  # here remove outlier from data.3 means third standard deviation
```

```python
sns.boxplot(x="CoapplicantIncome",data=df)
plt.show()
```

it is a scaling technique in which values are shifted & rescaled so that they end up ranging b/w  0 & 1.it is also known as Min-Max Scaling.(scaling dosn't effect outlier but its magnitude reduced).

# **Function Transformer**
convert non-normal distribution data into normal distribution data.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df=pd.read_csv("loan.csv")
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
sns.distplot(df["CoapplicantIncome"])
plt.show()
```

remove outlier using IQR

```python
q1=df["CoapplicantIncome"].quantile(0.25)
q3=df["CoapplicantIncome"].quantile(0.75)
IQR=q3-q1
```

```python
min_range=q1-(1.5*IQR)
max_range=q3+(1.5*IQR)
min_range,max_range
```

```python
df[df["CoapplicantIncome"]<max_range]
```

```python
df=df[df["CoapplicantIncome"]<max_range]
```

```python
sns.distplot(df["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft=FunctionTransformer(func=np.log1p)  # using log function for transform data
```

```python
ft.fit(df[["CoapplicantIncome"]])
```

```python
ft.transform(df[["CoapplicantIncome"]])
```

```python
df["CoapplicantIncome_Td"]=ft.transform(df[["CoapplicantIncome"]])
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(df["CoapplicantIncome"])
plt.title("Before")

plt.subplot(1,2,2)
sns.distplot(df["CoapplicantIncome_Td"])
plt.title("After")
plt.show()
```
